```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("StudentsPerformance.csv")
df
```

```python
df.describe()
```

```python
print(df[['math score', 'reading score', 'writing score']].describe())
```

```python
print(df[['math score', 'reading score', 'writing score']].mean())
```

```python
print(df[['math score', 'reading score', 'writing score']].median())
```

```python
print(df[['math score', 'reading score', 'writing score']].std())
```

```python
print(df[['math score', 'reading score', 'writing score']].max())
```

```python
print(df[['math score', 'reading score', 'writing score']].min())
```

```python
grouped=df.groupby('gender')['math score']
summery_stats=grouped.agg(['mean','median','min','max','std'])
print(summery_stats)
```

```python
male_scores = df[df['gender'] == 'male']['math score'].tolist()
female_scores = df[df['gender'] == 'female']['math score'].tolist()
```

```python
print("\nList of Male Math Scores:")
print(male_scores[:10]) 
```

```python
print("\nList of Female Math Scores:")
print(female_scores[:10])
```
