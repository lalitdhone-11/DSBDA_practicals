```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize, RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Ensure NLTK resources are downloaded (you can uncomment this if running for the first time)
nltk.download('all')

# 1. Read the sample document
content = """
Natural Language Processing (NLP) is a branch of Artificial Intelligence.
It helps computers understand human language.
NLP is used in chatbots, translation systems, and sentiment analysis.
Machine learning and deep learning are important in NLP.
"""

# 2. Tokenization
# Sentence Tokenization
sentences = sent_tokenize(content)
print("Sentence Tokens:", sentences)

# Word Tokenization
words = word_tokenize(content)
print("Word Tokens:", words)

# Custom Tokenization using Regular Expressions (e.g., words only)
tokenizer = RegexpTokenizer(r'\w+') 
custom_words = tokenizer.tokenize(content)
print("Custom Tokens:", custom_words)

# 3. Stop Word Removal
# Get a set of English stop words
stop_words = set(stopwords.words('english'))

filtered_words = []
for word in words:
    # Convert to lowercase and filter out stop words
    if word.lower() not in stop_words:
        filtered_words.append(word)
print("Words after Stop Word Removal:", filtered_words)

# 4. Part of Speech (POS) Tagging
# Apply POS tagging on the filtered words
pos_tags = nltk.pos_tag(filtered_words)
print("POS Tags:", pos_tags)

# 5. Stemming
# Initialize the Porter Stemmer
stemmer = PorterStemmer()
print("\n--- Stemming ---")
for word in words:
    print(word, "After stemming =", stemmer.stem(word))

# 6. Lemmatization
# Initialize the WordNet Lemmatizer
lemmatizer = WordNetLemmatizer()
print("\n--- Lemmatization ---")
for word in words:
    print(word, "After Lemmatization =", lemmatizer.lemmatize(word))

# 7. Term Frequency - Inverse Document Frequency (TF-IDF)
# Combine a few sentences into a single string within a list for the vectorizer
new_sentence = [" ".join(sentences[:3])]

def calculate_tfidf(document):
    # Initialize TF-IDF Vectorizer
    tfidf_vectorizer = TfidfVectorizer()
    
    # Fit and transform the document to create the matrix
    tfidf_matrix = tfidf_vectorizer.fit_transform(document)
    
    # Get feature (word) names
    feature_names = tfidf_vectorizer.get_feature_names_out()
    
    return tfidf_matrix, feature_names

# Call the function and print the matrix as an array
matrix, feature_names = calculate_tfidf(new_sentence)
print("\n--- TF-IDF ---")
print("Feature Names:", feature_names)
print("TF-IDF Matrix:", matrix.toarray())
```
